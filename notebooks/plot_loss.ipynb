```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/test_material/losses.csv')
df.head()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize = (16,6))

a = axes[0]
a.plot(df.train_loss, color = 'blue', label = 'Train Loss')
a.plot(df.test_loss, color = 'red', label = 'Test Loss')
a.grid(True)
a.legend(prop={'size': 16})
a.set_xlabel('Epochs', size=16)
a.tick_params(axis='both', which='major', labelsize=16)
a.tick_params(axis='both', which='major', labelsize=16)
a.set_title('Loss', size=20)

b = axes[1]

b.plot(df.train_auc, color = 'blue', label = 'Train ROC AUC')
b.plot(df.test_auc, color = 'red', label = 'Test ROC AUC')
b.grid(True)
b.legend(prop={'size': 16})
b.set_ylabel('Epochs', size=16)
b.tick_params(axis='both', which='major', labelsize=16)
b.tick_params(axis='both', which='major', labelsize=16)
b.set_title('ROC AUC', size=20)


plt.savefig('../doc/charts/loss_auc.pdf', format='pdf')
```
